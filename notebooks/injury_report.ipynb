# Cleaning Up the Injury Data Notebook

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loads in the data taken from FanGraph's RosterResource Injury Report
# https://www.fangraphs.com/roster-resource/injury-report?groupby=all&timeframe=all&season=2026

injury_data_list = []
for i in np.arange(2020, 2027):
    injury_data_list.append(pd.read_excel(f'../data/roster_resource_injury-report_{i}.xlsx'))
injury_data = pd.concat(injury_data_list, ignore_index=True)
injury_data = injury_data[~injury_data['MLBAMID'].isna()].reset_index(drop=True)
injury_data['MLBAMID'] = injury_data['MLBAMID'].astype(int)

## we filter to only include pitcher injuries
pitcher_pos = ['RP', 'SP', 'SP/RP', 'RP/SP']
injury_data = injury_data[injury_data['Pos'].isin(pitcher_pos)].reset_index(drop=True)

## normalizes team name
injury_data['Team'] = injury_data['Team'].apply(lambda x: 'ATH' if x == 'OAK' else x)

injury_data = injury_data[['MLBAMID', 'NameASCII', 
                           'Team', 'Pos', 
                           'Injury / Surgery', 'Injury / Surgery Date', 'Return Date', 
                           'Latest Update', 'Status'
                           ]]
injury_data
```
